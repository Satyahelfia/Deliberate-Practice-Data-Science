```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df

"""
.replace('N/a',''): Metode ini digunakan untuk mengganti setiap kemunculan string 'N/a' dengan string kosong (''). 
Hasilnya, nilai-nilai yang semula 'N/a' akan diubah menjadi string kosong.

.replace('NaN',''): Metode ini digunakan untuk mengganti setiap kemunculan string 'NaN' dengan string kosong (''). 
Hasilnya, nilai-nilai yang semula 'NaN' akan diubah menjadi string kosong.
"""
```

```python
df = df.fillna('')
df

"""
.fillna(''): Metode ini digunakan untuk menggantikan semua nilai yang bernilai NaN dengan nilai yang diberikan, dalam hal ini, string kosong (''). 
Hasilnya, nilai-nilai yang semula NaN akan diubah menjadi string kosong.
"""
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df

"""
.str.replace('Yes','Y'): Mengganti setiap kemunculan string 'Yes' dengan string 'Y'. 
Hasilnya, nilai-nilai yang semula 'Yes' akan diubah menjadi 'Y'.

.str.replace('No','N'): Mengganti setiap kemunculan string 'No' dengan string 'N'. 
Hasilnya, nilai-nilai yang semula 'No' akan diubah menjadi 'N'.
"""
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df

"""
for x in df.index:: Ini adalah loop for yang melakukan iterasi melalui indeks DataFrame df.

df.loc[x, "Do_Not_Contact"] == 'Y': Memeriksa apakah nilai dalam kolom "Do_Not_Contact" pada baris ke-x adalah 'Y'.

df.drop(x, inplace=True): Jika nilai 'Y' ditemukan, baris tersebut dihapus dari DataFrame df menggunakan metode drop. 
Parameter inplace=True menunjukkan bahwa perubahan ini akan diterapkan secara langsung pada DataFrame aslinya, tanpa perlu menugaskan hasilnya kembali ke variabel.
"""
```
